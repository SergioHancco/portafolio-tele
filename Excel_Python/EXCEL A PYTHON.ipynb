```python
import pandas as pd
import plotly.graph_objects as go
import matplotlib.pyplot as plt
df = pd.read_excel('Carreras_p.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
df['PROMEDIO'] = (df['NOTA 1'] + df['NOTA 2'] + df['NOTA 3'] + df['NOTA 4'] + df['NOTA 5'])/5
df
```

```python
df2=df[['ESPECIALIDAD','DEPARTAMENTO','NOTA 1','NOTA 2','NOTA 3','NOTA 3','NOTA 4','NOTA 5',"PROMEDIO"]]
df2
```

```python
df3=df2.groupby(['ESPECIALIDAD',df2['DEPARTAMENTO']]).mean()  
df3
```

```python
df3=df3.reset_index()
df3
```

```python
plt.figure(figsize=(15, 7), dpi =80)
plt.hist(df3['ESPECIALIDAD'], facecolor='purple', edgecolor='black', bins= 15)
plt.xticks(rotation=70)
plt.show()
```

```python
plt.figure(figsize=(15, 7), dpi =80)
plt.hist(df3['DEPARTAMENTO'], facecolor='purple', edgecolor='black', bins= 15)
plt.xticks(rotation=70)
plt.show()
```
